```python
lines = open('./metrics.csv').read().strip().split('\n')
```

```python
names=[[ t.strip('"') for t in line.split(',')] for line in lines[4:6]]
```

```python
for i,v in enumerate(names[0]):
    if not v:
        names[0][i]=names[0][i-1]
```

```python
names = list(zip(*names))
```

```python
values = [[float(t) for t in line.split(',')] for line in lines[6:]]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(values, columns=names)
```
